```python
%matplotlib widget

import matplotlib.pyplot as plt
from sim import build_simulations, SimulationType

def create_plot(name):
    fig, ax = plt.subplots()
    fig.set_label(name)
    ax.set_xlabel('Timesteps (sec)')
    ax.set_ylabel('State of charge (kilojoules)')
    return ax

for simulation in build_simulations():
    ax = create_plot(simulation.name)
    timesteps = range(0, len(simulation.charge_inputs) + 1)
    ax.plot(timesteps, simulation.simulate(SimulationType.COMPLETE), label='complete')
    ax.plot(timesteps, simulation.simulate(SimulationType.RELAXED), label='relaxed')
    ax.plot(timesteps, simulation.simulate(SimulationType.APPROXIMATE), label='approximate')
    ax.legend()
```
